# tensorboard

```python
import math
from torch.utils.tensorboard import SummaryWriter
```

```python
writer = SummaryWriter()

funcs = {"sin": math.sin,
         "cos": math.cos,
         "tan": math.tan}

for angle in range(-360, 360):
    angle_red = angle * math.pi / 180
    for name, fun in funcs.items():
        val = fun(angle_red)
        writer.add_scalar(name, val, angle)
writer.flush()
writer.close()
```
