## 한번에 여러 그래프 그리기

- stateless API (objected-based): 내가 지정한 figure, 내가 지정한 axes에 그림을 그리는 방법

- stateful API (state-based): 현재의 figure, 현재의 axes에 그림을 그리는 방법

```python
# stateless

fig, axes = plt.subplots(2, 1, figsize=(10, 6))

# first plot
axes[0].scatter(df['flipper_length_mm'], df['body_mass_g'])
axes[0].set_xlabel('날개 길이(mm)')
axes[0].set_ylabel('몸무게(g)')
axes[0].set_title('날개와 몸무게 간의 관계')

# second plot
axes[1].hist(df['body_mass_g'], bins=30)
axes[1].set_xlabel('Body Mass')
axes[1].set_ylabel('Count')
axes[1].set_title('펭귄의 몸무게 분포')

# 간격 조정
plt.subplots_adjust(
    left=0.1, right=0.95,
    bottom=0.1, top=0.95,
    wspace=0.5, hspace=0.5
)
plt.show()
```

```python
# stateful

plt.figure(figsize=(10, 6))

# first plot
plt.subplot(2, 1, 1)
plt.scatter(df['flipper_length_mm'], df['body_mass_g'])
plt.xlabel('날개 길이(cm)')
plt.ylabel('몸무게(g)')
plt.title('날개와 몸무게 간의 관계')

# second plot
plt.subplot(2, 1, 2)
plt.hist(df['body_mass_g'], bins=30)
plt.xlabel('Body Mass')
plt.ylabel('Count')
plt.title('펭귄의 몸무게 분포')

# 간격 조정
plt.subplots_adjust(
    left=0.1, right=0.95,
    bottom=0.1, top=0.95,
    wspace=0.5, hspace=0.5
)

plt.show()
```

```python
import seaborn as sns

df = sns.load_dataset('titanic')
df.head()
```

```python
# relation between age and fare
sns.scatterplot(data=df, x='age', y='fare')
plt.show()
```

```python
# hue : 그룹별 색
# style : 점의 색과 모양
sns.scatterplot(data=df, x='age', y='fare', hue='class', style='class')
plt.show()
```

```python
sns.heatmap(df_pivot, annot=True, cmap='coolwarm')
plt.show()
```

## 한 번에 여러 그래프 나타내기

- figure-level : matplotlib와 별개로 seaborn의 figure를 만들어 그곳에 그래프를 나타낸다. 따라서 figure-level 함수를 사용할 경우 facegrid를 통해 레이아웃을 변경하고, 여러 개의 그래프를 나타낼 수 있다.

- axes-level : matplotlib의 axes의 그래프를 나타낸다.

### figure level

```python
# 분포를 나타내는 그래프
# kind='hist' : 히스토그램
# hue : 'class'별로 색을 구분
# alpha : 투명도 조절
sns.displot(data=df, x='age', hue='class', kind='hist', alpha=0.3)
plt.show()
```

```python
# col : 특정 열을 입력하면 해당 열을 기준으로 그래프가 열 방향으로 각각 분할하여 표현
sns.displot(data=df, x='age', col='class', kind='hist')
plt.show()
```

```python
sns.displot(data=df, x='age', col='class', row='sex', kind='hist')
plt.show()
```

```python
# axes-level 함수인 histoplot()로는 facegrid를 할 수 없음
sns.histplot(data=df, x='age', col='class', row='sex')
```

### axes-level

```python
plt.rc('font', family='Malgun Gothic')

g, axes = plt.subplots(2, 1, figsize=(8, 6))    # 2개의 axes 생성

sns.histplot(data=df, x='age', hue='class', ax=axes[0])     # histplot() 함수를 통해 히스토함수 그린 후 ax=axes[0]을 입력하면 첫 번째 axes에 해당 그래프 출력
sns.barplot(data=df, x='class', y='age', ax=axes[1])        # barplot() 함수를 통해 막대 그래프를 그린 후, ax=axes[1]을 입력하면 두 번째 axes에 해당 그래프 출력

g.tight_layout()        # 여백 조정
plt.show()
```
