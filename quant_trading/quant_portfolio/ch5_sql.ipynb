## Sub queries

서브쿼리란 쿼리 내의 쿼리이며, 일회용 뷰를 의미한다. 즉 뷰를 정의하는 구문을 그대로 다른 구 안에 삽입하는 것이다. 뷰를 만든 후 이를 확인하는 쿼리를 다음과 같다.

```python
# creating a view and selecting data
# sql1 = """CREATE VIEW GoodSum (goods_classify, cnt_count) 
# AS
# SELECT goods_classify, COUNT(*)
# FROM goods
# GROUP BY goods_classify
# """

# sql2 = """
# SELECT * FROM GoodSum
# """

sql = """
SELECT goods_classify, cnt_goods
FROM (
    SELECT goods_classify, COUNT(*) AS cnt_goods
    FROM goods
    GROUP BY goods_classify
) AS GoodSum;
"""

select_sql(sql)
```

## Scalar sub-queries

스칼라 서브쿼리란 단일 값이 반환되는 서브쿼리다. 이를 통해 =, <, > 등 비교 연산자의 입력값으로 사용할 수 있다. 예를 들어, 판매단가가 전체 평균 판매단가보다 높은 상품만을 검색하려는 경우

```python
sql = """
SELECT *
FROM goods
WHERE sell_price > (SELECT AVG(sell_price) FROM goods);
"""

select_sql(sql)
```

```python
sql = """ 
SELECT goods_id, goods_name, sell_price, 
    (SELECT AVG(sell_price) FROM goods) AS avg_price
FROM goods
"""

select_sql(sql)
```

## window function

랭킹, 순번 생성 등 일반적인 집약 함수로는 불가능한 고급 처리 가능

<윈도우 함수> over ([partition by <열 리스트>] order by <정렬용 열 리스트>)

```python
# rank function
sql = """
SELECT goods_name, goods_classify, sell_price, 
    RANK() OVER (PARTITION BY goods_classify ORDER BY sell_price) AS ranking
FROM goods;
"""
select_sql(sql)
```

1. partition by 는 순위를 정할 대상 범위를 설정하며, 어떤 조건으로 그룹을 나눈다고 생각하면 이해가 쉽다. 상품 분류마다 순위를 구하고자 하므로 goods_classify를 입력한다.

2. order by는 윈도우 함수를 어떤 열에 어떤 순서로 적용할지 정한다. 판매단가를 오름차순으로 순위를 구하고자 하므로 sell_price를 입력하였다.

3. 순위를 구하는 윈도우 전용 함수인 rank()를 입력한다.


이 중 partition by를 통해 구분된 레코드 집합을 '윈도우'라고 하며, 이는 '범위'를 타나낸다. 만일 partition by를 지정하지 않으면 전체 테이블이 윈도우가 되므로, 아래와 같이 sell_price 열 자체를 기준으로 순위가 구해진다.

```python
sql = """ 
SELECT goods_name, goods_classify, sell_price,
    RANK() OVER (ORDER BY sell_price) AS ranking
FROM goods
"""
select_sql(sql)
```

```python
# various ranking
sql = """ 
SELECT goods_name, goods_classify, sell_price,
    RANK() OVER (ORDER BY sell_price) AS ranking,
    DENSE_RANK() OVER (ORDER BY sell_price) AS dense_ranking,
    ROW_NUMBER() OVER (ORDER BY sell_price) AS row_ranking
FROM goods;
"""

select_sql(sql)
```

```python
# an aggregation function as a window function
# over를 빈 칸으로 둘 경우 current_sum 열에는 모든 sell_price의 합계가 나타난다.
sql = """ 
SELECT goods_id, goods_name, sell_price,
    SUM(sell_price) OVER() as current_sum
FROM goods
"""

select_sql(sql)
```

```python
# 누적합계
sql = """ 
SELECT goods_id, goods_name, sell_price,
    SUM(sell_price) OVER(ORDER BY goods_id) AS current_sum
FROM goods;
"""

select_sql(sql)
```

```python
# 누적 평균
sql = """ 
SELECT goods_id, goods_name, sell_price,
    AVG(sell_price) OVER(ORDER BY goods_id) AS current_avg
FROM goods;
"""
select_sql(sql)
```

```python
# partition by 추가해서 윈도우별로 집계
sql = """ 
SELECT goods_id, goods_classify, goods_name, sell_price,
    SUM(sell_price) OVER(PARTITION BY goods_classify ORDER BY goods_id) AS current_sum
FROM goods;
"""

select_sql(sql)
```

### Moving average

윈도우 함수에서는 그 범위를 정해 '프레임'을 만들 수도 있다. 이는 over 내의 order by 구문 뒤에 범위 지정 키워드를 사용하면 된다. 예를 들어, 최근 3개 데이터만 이용해 평균을 구하는 이동평균을 계산하는 쿼리는 다음과 같다.
rows n preceding을 입력할 경우 앞의 n행까지만 프레임을 만들어 계산한다. 예제에서는 n=2를 입력했으므로, 현재 행과 앞의 2개 행, 즉 3개 행으로만 이동평균을 계산한다.

```python
sql = """ 
SELECT goods_id, goods_classify, goods_name, sell_price,
    AVG(sell_price) OVER(ORDER BY goods_id ROWS 2 PRECEDING) AS moving_avg
FROM goods
"""
select_sql(sql)
```

앞의 행이 아닌 뒤의 행을 이용해 계산하고 싶을 경우 preceding 대신 following을 입력한다. 현재 행과 뒤의 2개 행으로 이동평균을 계산하는 방법은 다음과 같다.

```python
sql = """ 
SELECT goods_id, goods_classify, goods_name, sell_price,
    AVG(sell_price) OVER(ORDER BY goods_id ROWS BETWEEN CURRENT ROW AND 2 FOLLOWING) AS moving_avg
FROM goods;
"""

select_sql(sql)
```

```python
# following과 preceding 동시 사용
sql = """ 
SELECT goods_id, goods_classify, goods_name, sell_price,
    AVG(sell_price) OVER(ORDER BY goods_id ROWS BETWEEN 1 PRECEDING AND 1 FOLLOWING) AS moving_avg
FROM goods;
"""

select_sql(sql)
```
